```python
import matplotlib.pyplot as plt
```

```python
file.fillna(0, inplace=True)
file
```

```python
fecha = file["Datetime"].head(1000)
elevation = file["Elevation (deg)"].head(1000)
ext_radiation = file["Extraterrestrial Radiation (W/m2)"].head(1000)
pdc = file["Pdc (W)"].head(1000)
pac = file["Pac (W)"].head(1000)
```

```python
%matplotlib inline

# Crear las tres gráficas separadas
plt.figure(figsize=(10, 6))  # Tamaño de la figura

# Primera gráfica
plt.subplot(1, 3, 1)  # 3 filas, 1 columna, posición 1
plt.bar(fecha, pac, label='pdc')
plt.title('Behaviuor 1')
plt.legend()
plt.xlabel("Fechas")
plt.ylabel("Pac (W)")

# Segunda gráfica
plt.subplot(1, 3, 2)  # 3 filas, 1 columna, posición 2
plt.plot(pdc, pac, label='line')
plt.title('Behaviour 2')
plt.legend()
plt.xlabel("Pdc (W)")
plt.ylabel("Pac (W)")

# Tercera gráfica
plt.subplot(1, 3, 3)  # 3 filas, 1 columna, posición 3
plt.scatter(elevation, ext_radiation, label='points')
plt.title('Behaviour 3')
plt.legend()
plt.xlabel("Elevation (deg)")
plt.ylabel("ext_radiation (W/m2)")

# Ajustar espaciado entre las gráficas
plt.tight_layout()
plt.savefig('./bar_ver.png')
# Mostrar las gráficas
plt.show()
```

```python
#file_name = './data/SD_5Min.xlsx' # ./ es pwd()
file_name = "/content/sample_data/SD_5Min.xlsx"
sheet = 'Radiacion'
df = pd.read_excel(io=file_name, sheet_name = sheet, engine='openpyxl')
```

```python
df.drop('Lufft Irrad (W/m2)', axis=1, inplace=True)
```

```python
df
```

```python
fecha = '2020-03-28'
marzo = df[(df['Date'].dt.month == pd.to_datetime(fecha).month)&(df['Date'].dt.year == pd.to_datetime(fecha).year)]
```

```python
un_dia = marzo[marzo["Date"].dt.day == pd.to_datetime(fecha).day]
tiempos = pd.DatetimeIndex(un_dia["Date"]).time
```

```python
dict_ = {}
for i in range(1,32):
  marzo_ = marzo[marzo["Date"].dt.day == pd.to_datetime('2020-03-{0}'.format(i)).day]["Meteocontrol Irrad (W/m2)"].to_list()
  dict_["Irrad Dia {0}".format(i)] = marzo_
```

```python
pd_f = pd.DataFrame(dict_)
medias = pd_f.mean(axis=1)
min = pd_f.min(axis=1)
max = pd_f.max(axis=1)
```

```python
pd_f.insert(0,"Tiempo",tiempos)
```

```python
pd_f["Min"] = min
pd_f["Prom"] = medias
pd_f["Max"] = max
```
